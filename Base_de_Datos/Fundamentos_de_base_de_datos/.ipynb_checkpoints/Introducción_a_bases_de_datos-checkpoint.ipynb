# Introducción a base de datos

## Entidades y atributos:

- Entidad (se encierran en rectangulos): Una entidad es algo muy similar aun objeto, seria como que representa algo en la vida real, como ser algo abstracto, por ejemplo: Automoviles, Computadoras, Personas, Hospitales, etc. --> Por convencion vamos a escribir en plural las entidades, ya que representan un conjunto/grupo de objetos.

- Atributos (se encierran en ovalos): Son como las propiedades de la entidad, por ejemplo, para la entidad "Automóvil", un atributo del mismo podria ser: "Color", "Fecha_De_Fabricacion", etc. En este caso de Entidad, un atributo en particular como ser "llantas" (al ser un automovil, el mismo tendra más de una llanta) se encierran en un doble óvalo, debido a que es un atributo multivaluado (por lo que tiene multiples valores).

- Atributos Compuestos: siguiendo el caso en donde mi entidad es un "automovil",un Atributo más del mismo podria ser el "motor", sin embargo el motor esta compuesto por otros atributos como ser "pistones", "bugias", etc. es por ello que a este tipo de atributos se le llaman Atributos compuestos.

- Atributo especial del caso en donde mi entidad es el "Automovil", donde si propongo como un atributo la "antiguedad" la misma debemos de inferirla, por ello este tipo de atributos especiales las encerramos en `óvalo punteado`.

Donde se puede observar tres casos particulares de entidades:

- `disco duro`: Encerrado en dos óvalos, donde indica que puede tener más de un valor, por ejemplo una lapto puede tener más de un disco duro.
- `antiguedad`: Encerrado por un óvalo punteado, donde indica que el valor de ese atributo habria que inferirlo.
- `métodos de entrada`: donde se puede observar que se trata de un Atributo Compuesto, esto debido a que dicho atributo esta compuesto por más atributos.
- **`no de serie`**: es un atributo cuyo valor es unico, y nos va a permitir identificar una computadora de las demas, este tipo de atributo que se convierte en lo que se llama `llave primaria`, se escribe una linea debajo del nombre del atributo que diferencialos de los demas atributos, siendo está la `llave primaria`.

## Cardinalidad

### Cardinalidad 1:1

<img src=".\img\9.jpg" width="400px" height="400px">
Analisis:

- Una persona, Cuantos datos de contactos tiene?, Tiene solo 1 datos de contacto.
- Un dato de contacto, a cuantas personas puede pertenecer?, Solo a una persona.

**IMPORTANTE**: Para sacar la cardinalidad entre estas dos entidades lo que debemos hacer es tomar el numero mayor de ambos lados, en este caso el mayor de ambos lados es el 1, por ello la cardinalidad es 1:1.

<img src=".\img\10.jpg" width="400px" height="400px">

### Cardinalidad: 0 a 1

Tambien se le suele llamar Cardinalidad 1 a 1 opcional, lo que significa que puede haber la opcion que exista uno o ninguno de los lados.

<img src=".\img\11.jpg" width="400px" height="400px">

Analisis:

- Una sesion actual tiene que tener un usuario.
- Pero un usuario puede no estar en sesion en este momento

Por lo que la cardinalidad se define de 0 a 1.


### Cardinalidad: 1 a N

O tambien llamada 1 a muchos. 

Los que quiere decir, como su nombre lo indica, que de un lado podemos tener 1 uno y del otro lado podemos tener muchos.
<img src=".\img\12.jpg" width="400px" height="400px">
Analisis:
-  Una persona puede tener muchos automoviles.
- Sin embargo un automovil solo puede pertenecer a una persona (legalmente, por papeles solo puede tener 1 dueño)

Donde para tomar la cardinalidad de esta relacion debo tomar el mayor valor de ambos extremos, en este caso 1:N.

### Cardinalidad: 0 a N

Tambien conocido como 0 a N opcional.
<img src=".\img\14.jpg" width="400px" height="400px">

**Texto:**
- CHAR(n): Me permite almacenar tanto caracteres, como cadenas. (si guardaba en memoria espacio para una cadena de 8 caracteres, solo se podrian usar estos 8 caractenes declarados).
- VARCHAR(n): Tambien me sirven para almacenar cadenas, con la diferencia que el tipo de dato CHAR toma un pedazito de memoria del número de caracteres que voy a guardar, mientras que VARCHAR hace esto de forma más dinamica, ya que reservada un minimo de espacio en memoria y a partir de ahi iba creciendo o encogiendose hasta el maximo que declarara, para permitir guardar cadenas mas grandes o más chicas. (es más eficiente para cuando no se sabia la logitud de la cadena). (limite de caracteres de 255)
- TEXT: Tambien sirve para almacenar cadenas, pero más que nada para almacenar cadenas de caracteres muy grandes.

**Números:**

- INTEGER: Entero, el numero que no tiene punto decimal o no tiene fraccion. Nos sirve para cuando necesito hacer una multiplicacion, suma, resta con este tipo de dato.
- BIGINT: Cuando necesitamos almacenar datos númericos muy grandes.
- SMALLINT: Cuando sabemos que el numero sera 99 o menos, podemos declararlos con este tipo de dato.
- DECIMAL(n,s): Necesita dos parametros de entrada, en "n" es que voy a añadir ese números, y "s" que voy a agregar esos decimales. Sirve cuando deseo ingresar numeros con una mayor presicion.
- NUMERIC(n,s): Necesita dos parametros de entrada, en "n" es que voy a añadir ese números, y "s" que voy a agregar esos decimales. Sirve cuando deseo ingresar numeros con una mayor presicion.

**Fechas/hora:**

- DATE: solo contiene la fecha (año, mes y dia)
- TIME: que contiene la hora del dia (las 24 horas).
-  DATETIME: Contiene tanto el dia como la hora, incluso en milisegundos. (sirve para cuando quiero guardar algo con mucha presicion).
- TIMESTAMP: Contiene tanto el dia como la hora, incluso en milisegundos. (sirve para cuando quiero guardar algo con mucha presicion).

**Lógicos:**

- BOOLEAN: solo puede tener dos valores (es un tipo de dato binario, puede ser un 1 o un 0). Suele usarse para hacer un validacion con una bandera.

#### Trabajar con cadinalidades n a n (casos especiales):

Analisis entre post y etiquetas
<img src=".\img\27.jpg" width="400px" height="400px">

Cuando esto sucede no podemos asignar el id de referencia para la FK como en los anteriores casos. En este tipo de casos debemos crear una nueva tabla, llamada "tabla pivote", para ayudar a ver cual es la relacion entre estas entidades. Esta nueva tabla generalmente suele llevar el nombre de la union entre las dos tablas de ambas entidades con cadinalidad n a n. Esta nueva tabla tendra el id de ambas tablas.


**Clave compuesta** (se usara en la tabla pivote "posts_etiquetas"): Es cuando tenemos el campo "id", pero no es solo un campo, en este caso estaria formado por dos campos (post_id, etiqueta_id). Donde la combinacion entre el post_id, con el etiqueta_id debe ser unica. Tambien se puede fabricar una id para identificar un registro, como ser posts_etiquetas_id.
