# Medidas de  dispersión

## Dispersión en una distribución

<img src=".\img\Estadistica_Descriptiva\5.png" width="600px" height="600px">

- **Rango:** Es el valor de restar el valor maximo y el valor minimo pertenecientes al conjunto de datos.
- **Rango intercuartil:** ienen como propiosito definir mi conjunto de datos en termino de 4 subdivisiones homogeneas (homogeneas: tener la misma cantidad de datos en cada subdivision.)
    - **Q2:** (Primer cuartil -> Mediana). Esta dado por la mediana, ya que es el dato que parte en la mitad los datos.
    - **Q1:** Viene dado luego de partir a la mitad la porcion de la parte izquierda de Q2. En otras palabras Q1 es la mitad entre la mediana y el valor minimo.
    - **Q3:** Es el valor dado por la mitad entre la mediana (Q2) y el valor maximo.
    
    - **Rango Intercuartil:** Es la distancia entre el cuartil tres (Q3) y el cuartil uno (Q1).
- **Desviación estándar**


## Desviación estándar:

<img src=".\img\Estadistica_Descriptiva\6.png" width="800px" height="800px">

Para una muestra se divide en $n-1$ mientras que para una poblacion solo se divide en $n$.


### Distribucion Normal:

<img src=".\img\Estadistica_Descriptiva\7.png" width="800px" height="800px">

Tiene forma de la campana de Gauss, la cual es simetrica respecto al Q2 (mediana). Los brazos que salen a la caja se les puede asociar al maximo o al minimo, pero en realidad se suelen ajustar las barras (circunferencia verde) para que nos permitan excluir lo que llamamos los datos anomalos o "outliers" (como son tambien llamados).

Para tratar con los datos anomalos se suele usar la **desviación estándar**, en este caso al tratarse de una distribucion normal, la media ($mu$) y mediana coinciden en cuando su valor. Cuando tenemos datos anomalos en el conjunto de nuestros datos se suele tomar el **99.72%** de los mismos, estos datos estaran contenidos entre $media - 3 * desviación_estándar$ y $media + 3 * desviación_estándar$, tomando de esta forma todos los valores que se encuentran entre ese rango de valores, mientras que los valores anomalos son los valores que se encuentran en la cola de la distribucion y no son tomados en cuenta.

**Mapear en IQR**

Como mencione arriba los brazos no haces referencia exactamente a los valores maximos y minimo el conjunto de datos, sino que estos vienen datos por un calculo. Los valores anomalos(en imagen **X**) son aquellos que quedan fuera de las barras (brazos de la caja), que normalemente en analisis estadisticos, son datos que se descartan. El calculo para encontrar los verdaderos valor de `max` y `min` se llama metódo de deteccion de outliers con el rango intercartil.

El **1.5**, es el valor que cuando lo correspondemos con los limites de la desviacion estandar ($media - 3 * desviación_estándar$ y $media + 3 * desviación_estándar$), coindice en el hecho de cuando estariamos contemplando la mayoria de los datos, alrededor del 99.72%.


### Distribucion Sesgada:

<img src=".\img\Estadistica_Descriptiva\8.png" width="500px" height="500px">

El corchete en rojo para calculo de las barras es erroneo, la verdadera forma de calcular el largo de las barras es usando una funcion cualquiera, que dependa del rango intercuartil. La idea de colocar las funcion $f$ y $g$ es dar una idea del sesgo de la distribucion, de esa asimetria de la distribucion y que entonces se logre calcular la longitud de la barras, de forma que se ajusta perfectamente a la distribucion real.

```python
iqr = Q3 - Q1  # Rango intercuartil
iqr
```

# Pipelines de procesamiento de datos numéricos

## Escalamiento lineal

### Por que usarlos?
    - Modelos de machine learning eficientes en el rango [-1,1].
    
### Hay diferentes tipos?
    - mix-max, Clipping, Z-Score, Winsorizing, etc.

### Cuando usarlos?
    - Data simétrica o uniformemente distribuida
    
<img src=".\img\Estadistica_Descriptiva\10.png" width="700px" height="700px">

Para que el algoritmo de ML pueda optimizar los datos, dichos datos deben estar entre el rango de [-1,1]. Para lograr eso, usarmos el **Escalamiento**:

### **Escalamiento min-max:** 
Cada dato que tenga $X$ (valor sin normalizar), lo debo normalizar $X_s$ (X scale o X escalado), mediante una transformacion, de forma que ahora los valores normalizados se encuentren entre [-1,1].

Formula de Transformacion:  $X_s = (2X-min-max) / (max-min)$ Donde el min y max son los valores correspondientes al conjunto de datos.


### **Escalamiento Clipping: (No es comun usar este metódo en ML)** 
Trata de tomar la distribucion, y con la distribucion, cortaremos todos los datos entre valore limites (Limite inferior $(-)$ y limite superior $(+)$), asumiendo que todos los datos entre esos limites (Franja roja), siguien siendo exactamente iguales, mientras que los datos que queden fuera (franja verde), se veran colapsados.

Ejemplo:

En caso de tener en valor 3.8 dentro de el rango limite, un valor 6.0 colapsado del lado derecho del limite superior y un valor 1.0 colapsado en el lado izquierdo del limite inferior.

$X_S$  -->  3.8  -->  3.8  # Queda exactamente igual, ya que se encuentra en la franja verde.
            6.0  -->  4.0  # Lo que se hace con el 6.0 es colapsarlo al limite superior.
            1.0  -->  2.0  # Lo que se hace con el 1.0 es colapsarlo al limite inferior.
            
Lo que se esta haciendo es forzar a todos los datos colapsados a quedar en los extremos de la distribucion.


**Nota:** Los valores Max y Min se pueden escoger de forma arbitraria, o se pueden elegir en base a los percentiles. Cuando se toma en cuenta los percentiles para los valores de Max y Min, es cuando el metodo de clippping pasa a llamarse **Metódo de Winsorizing**


### **Metódo Z-Score:** (Es el más comun)
Es el metódo usado más comun, debido a que dicho metódo esta basado precisamente en la definicion de promedio y desviacion estándar.

<img src=".\img\Estadistica_Descriptiva\11.png" width="500px" height="500px">

Contando con un conjunto de datos, obtengo la media y desvicion estandar, y contando con que tengo una distribucion simetrica, entonces los valores normalizados se calculan como, (el valor original - media)/desviacion estandar.

Formula de normalizacion:  $X_s = (X - \mu) / \sigma$

De esta forma, estaria garantizando, que lo datos van a estar normalizados dentro del rango razonable, aunque es probable que hallan algunos numeros que queden por fuera, pero las desviacion no son tan extremas.

Lo que se hace al normalizar los datos es, mantener la distribucion gaussiana (simetrica) con promedio 0 y una desviacion estandar de 1. De esa forma hacemos que la gran mayoria de los datos se ajusten al rango ([-1,1]) ideal para los algortimos de machine learning.

# Transformación no linea

## Escalamiento no lineales

### Por que usarlos?
    - Datos fuertemente sesgados, no simétricos.
    
### Hay diferentes tipos?
    - Logaritmos, sigmoides, polinomiales, etc.

### Cuando usarlos?
    - Antes de escalamientos lineales!.
    
**Procedimiento para tratar con datos cuya distribucion es no simetrica:**

<img src=".\img\Estadistica_Descriptiva\12.png" width="600px" height="600px">

Para tratar con datos estos tipos de datos, primero lo que se hace es aplicar una transformacion, de forma que los datos obtenido de dicha transformacion le apliquemos el escalonamiento lineal.

La transformacion sera hecha atravez de $X_s = tangh(x)$, donde $x$ son los valores de datos con distribucion sesgada, se usa la funcion $tangh()$ debido a que el rango de dicha funcion esta entre $[-1,1]$, que es el estandar para la optimizacion en los algoritmos de ML.

Lo que obtenemos luego de la transformacion de los datos es una distribucion casi simetrica (Curva Morada), que es lo que se necesita para aplicar el **Escalamiento Lineal**.

### Calibrar parametro de simetrizacion:

Es posible modificar la curva morada, para que la distribucion sea aún más simetrica, y eso se logra con el parametro $a$, donde, dandole diferentes valores puedo lograr que la curva de $tangh(x/a)$ sea más suave o más pronunciada, afectando de esa forma los valores transformados y asi la nueva curva.

# Procesamiento de datos numéricos en Python

## Escalenamiento de datos numéricos

### Transformaciones lineales

```python
import timeit  # Medir performans del modelo. (tiempo de ejecucion de la ejecucion)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model

X, y = datasets.load_diabetes(return_X_y = True)
raw = X[:,None, 2]
```

```python
# Reglas de escalamiento

max_raw = max(raw)
min_raw = min(raw)
scaled = (2*raw - max_raw - min_raw)/(max_raw - min_raw)

# Grafica datos originales vs Datos escalonados

fig, axs = plt.subplots(2, 1, sharex = True)

axs[0].hist(raw)
axs[1].hist(scaled)
```

```python
# Modelos para entrenamiento

def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled, y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_time = timeit.timeit(train_scaled, number=100)

print('train raw : {}'.format(raw_time))
print('train scaled : {}'.format(scaled_time))
```

- max-min scaling: Mejor para datos uniformemente distribuidos.
- Z-score scaling: Mejor para datos distribuidos "normalmente" (forma de campana de gauss)

# Correlaciones

Si grafico dos variables en un scatterplot y veo cuando una varia en la misma direccion que la otra de una forma muy bien definida, se podria decir que estan muy relacionadas, (de una forma informal decimos que estan correlacionadas), entonces no tiene sentido incluir ambas variables en un analisis de un modelo de ML, por que probablemente ambas variables esten aportando la misma informacion si su correlacion es muy alta, por lo cual deberiamos eliminar alguna de esas dos.

# Varianza

La Varianza de un conjunto de datos (que representan una misma variable, en este caso, la variable x), viene calculada por la suma cuadratica de las desviaciones que cada dato tiene respecto al valor promedio des esa variable x , y dividimos por $n-1$ (longitud de las observaciones que tenemos para es variable).

<img src=".\img\Estadistica_Descriptiva\17.png" width="400px" height="400px">

# Covarianza

Es la varianza conjunta entre dos variables.

<img src=".\img\Estadistica_Descriptiva\18.png" width="400px" height="400px">


## Estandarizar las variables a una misma escala: Coeficiente de Correlacion

Sucede que al analizar la relacion entre dos variables, es posible que ambas esten en unidades diferentes y en diferentes ordenes, ya que es como si hablaramos de relacion "peras y manzanas", es por ello que para estandarizar esto, se usa la formula de covarianza y que cada una de las desviaciones de las variables, las dividamos por la desviacion estandar, para estandarizar las variables a una misma escala. Esto conduce a un concepto que se llama el coeficiente de correlacion (ro --> P), que se calcula como la $Cov$ / ($std(x)$*$std(y)$.

<img src=".\img\Estadistica_Descriptiva\15.png" width="300px" height="300px">

**Importante:** Cuando tenemos un coeficiente de correlacion muy alto, es porque las variables tiene una correlacion elevada, es cuando es muy baja las variables tiene una correlacion muy baja.

## Covarianza y coeficiente de correlación

<img src=".\img\Estadistica_Descriptiva\19.png" width="800px" height="800px">

**Grafico Izquierdo:**

Se nos muestra datos que estan poco relacionados, por que las variaciones en el eje x, no muestran una relacion clara con las variaciones en el eje y, y alli vemos que el coeficiente de correlacion es 0.269, que esta muy cercano al 0.

**Grafico Derecho:**

Vemos que los datos tienen una relacion mucho más clara, se puede ver que a medida que aumenta el valor de la variable x, de la misma forma aumenta el valor del dato de la variable y. Y se puede observar que el coeficiente de correlacion es 0.962, que esta muy cercano al 1.


### Analizis del valor del coeficiente de correlación:

Luego de observar los graficos, se puede interpretar de valor del coeficiente de correlación, que a medida que el valor este cercano 0, nos dice que los datos no tienen absolutamente ninguna correlacion, mientras que si el dato es cercano a 1, tiene una correlacion directa.

**Observacion:** Si obtengo un valor para el coeficiente de correlación igual a -1, me dicen que los datos tienen una correlacion, pero la misma es inversa, es decir que mientras que una variable aumenta, la otra disminuye.

**Conclusion: El coeficiente de correlación es un valor que se encuentra entre -1 y 1, mientras más cercano este a esos valor, existe una gran correlacion entre los datos, mientras que por lo contrario, si el valor obtenido es 0 o cercano, se puede decir que los datos no tiene ninguna correlacion entre esas variables.**



### Correlacion no implica causacion.

La correlacion es una cantidad que simplemente mide, si una por casualidad o no, una variable al variar en una direccion, la otra tambien lo hace en la misma direccion, esto seria solo una relacion estadistica. Pero Cuando se mira el transfondo del problema, no quiere decir que halla una relacion causa y efecto entre una variable y la otra.

--> "Causacion no esta osociado con correlacion.", o mejor dicho --> "Correlacion no implica Causacion."

Es importante tener en claro que cuando tenga variables altamente correlacionadas en mi conjunto de datos, no quiere decir que una sea necesariamente cuasa de la otra (puede que si, pero no necesariamente).

# Reducción de dimensionalidad con PCA: En Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')

scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
)

covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
sns.pairplot(iris)
```

```python
sns.jointplot(x = iris['petal_length'], y = iris['petal_width'])
sns.jointplot(x = scaled[:,2], y = scaled[:,3])
```

```python
# Calcular descomposicion en vectores y valores propios de una matriz de forma inmediata

eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

### Cada vector propio es una de las direcciones principales a lo largo de la cual capturamos varianza de los datos originales.

```python
variance_explained = []

for i in eigen_values:
    variance_explained.append((i/sum(eigen_values))*100)
    
print(variance_explained)

# La primera componente, por si sola ya captura el 72% de la varianza total de las 4 dimenciones.
```

## PCA con scikit:

Como ponemos hacer la transformacion, para tomar esos datos que estan en 4 dimensiones y reducirlos a menos dimensiones, en base en esto.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca.fit(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
# Reduccion de los datos

reduced_scaled = pca.transform(scaled)
```

```python
reduced_scaled
```

```python
iris['pca_1'] = scaled[:,0]
iris['pca_2'] = scaled[:,1]

iris
```

```python
sns.jointplot(iris['pca_1'],iris['pca_2'], hue = iris['species'])
```
